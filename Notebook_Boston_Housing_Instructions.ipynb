# Boston Housing Regression Exercise 


## Scenario
You are working as a **Junior Data Analyst / Data Scientist** at a real-estate analytics company.
Your task is to:
- Understand what influences **median house prices** (`medv`).
- Determine which variables are most important.
- Explain why correlation ≠ regression importance.

You must analyze the predictors:
- `rm` – average number of rooms
- `lstat` – % lower-status population
- `ptratio` – pupil–teacher ratio

## Deliverables
### 1. Dataset Summary
Write a short overview including:
- Number of rows & columns
- What is the target variable (`medv`)?
- What the predictors (`rm`, `lstat`, `ptratio`) represent.

### Exploratory Data Analysis (EDA)
Provide:
- Short text explanation of each variable's distribution
- Interpretation of correlations:
  - `rm` vs `medv`
  - `lstat` vs `medv`
  - `ptratio` vs `medv`

### Regression Models to Create
Build **4 models** and report:
- R²
- Adjusted R²
- RSS (Residual Sum of Squares)

Models:
1. Full model (`rm + lstat + ptratio`)
2. Without `rm`
3. Without `lstat`
4. Without `ptratio`

### Final Interpretation
A short explanation (5–10 sentences):
- Which variable is most important?
- Why removing `lstat` drops R² more than removing `rm`?
- What this says about multicollinearity?
- Why correlation ≠ unique importance?

## Final Output Requirements
Submit a short written report containing:
- Scenario
- EDA summary
- Model metrics
- Comparison table
- Final interpretation paragraph
